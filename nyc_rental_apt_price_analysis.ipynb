<h1 align="center">NYC Rental Apartment Price Analysis</h1>

**Context & Objective**: Given NYC rental apartment data from Jan. 2012 to Sept. 2021, I've investigated the following questions using various Python data science libraries:
1. [What do rental apartment prices look like from the pre-pandemic high to Sept. 2021?](#first_link)
2. [What do rental apartment prices look like from the pre-pandemic high to the pandemic low?](#second_link)
3. [What are currently the most expensive neighborhoods for a studio apartment in Manhattan and Brooklyn?](#third_link)
4. [What have apartment prices looked like from Jan. 2012 to Sept. 2021?](#fourth_link)
5. [What have apartment prices looked like from Jan. 2020 to Sept. 2021?](#fifth_link)

*Note: This analysis was conducted in Oct. 2021.*

**Data Source**: The data was provided in a csv file directly from StreetEasy, a real estate listing platform for the New York metropolitan area.

**Data Methodology**: The StreetEasy Rent Indices are monthly indices that track changes in rent for all housing types and are currently available from January 2007 in Manhattan, January 2010 in Brooklyn and January 2012 in Queens. Each index uses a repeat-sales method similar that used to calculate the StreetEasy Price Indices. The repeat method evaluates rental price growth based on homes in a given geography that have listed for rent more than once.

## What do rental apartment prices look like from the pre-pandemic high to Sept. 2021?<a id='first_link'></a>
- Code below, then visualization presented
- *This analysis was conducted in Oct. 2021.*

```python
pandemic_df = df.loc['2019-01-01':'2021-09-01']

# "Fake" Data to set x points
nyc_date_x = [1, 2]
manhattan_date_x = [4, 5]
brooklyn_date_x = [7, 8]
queens_date_x = [10, 11]

nyc_prices_y = [pandemic_df['NYC'].max(), pandemic_df['NYC'].iloc[-1]]
manhattan_prices_y = [pandemic_df['Manhattan'].max(), pandemic_df['Manhattan'].iloc[-1]]
brooklyn_prices_y = [pandemic_df['Brooklyn'].max(), pandemic_df['Brooklyn'].iloc[-1]]
queens_prices_y = [pandemic_df['Queens'].max(), pandemic_df['Queens'].iloc[-1]]

plt.figure(figsize=(10, 6))

nyc_bar = plt.bar(nyc_date_x, nyc_prices_y, label='NYC', color='green', width=0.4)
manhattan_bar = plt.bar(manhattan_date_x, manhattan_prices_y, label='Manhattan', color='#0000FF', width=0.4)
brooklyn_bar = plt.bar(brooklyn_date_x, brooklyn_prices_y, label='Brooklyn', color='#3333FF', width=0.4)
queens_bar = plt.bar(queens_date_x, queens_prices_y, label='Queens', color='#6666FF', width=0.4)

plt.plot(nyc_date_x, nyc_prices_y, 'red', marker='o')
plt.plot(manhattan_date_x, manhattan_prices_y, 'red', marker='o')
plt.plot(brooklyn_date_x, brooklyn_prices_y, 'red', marker='o')
plt.plot(queens_date_x, queens_prices_y, 'red', marker='o')

nyc_change = '- ' + str(round((pandemic_df['NYC'].iloc[-1] - pandemic_df['NYC'].max()) / pandemic_df['NYC'].max() * -100, 1)) + '%'
manhattan_change = '- ' + str(round((pandemic_df['Manhattan'].iloc[-1] - pandemic_df['Manhattan'].max()) / pandemic_df['Manhattan'].max() * -100, 1)) + '%'
brooklyn_change = '- ' + str(round((pandemic_df['Brooklyn'].iloc[-1] - pandemic_df['Brooklyn'].max()) / pandemic_df['Brooklyn'].max() * -100, 1)) + '%'
queens_change = '- ' + str(round((pandemic_df['Queens'].iloc[-1] - pandemic_df['Queens'].max()) / pandemic_df['Queens'].max() * -100, 1)) + '%'

plt.text(nyc_bar[0].get_x() + 0.1, nyc_bar[0].get_height(), nyc_change, fontdict={'fontsize': 13, 'color': 'red'})
plt.text(manhattan_bar[0].get_x() + 0.1, manhattan_bar[0].get_height(), manhattan_change, fontdict={'fontsize': 13, 'color': 'red'})
plt.text(brooklyn_bar[0].get_x() + 0.1, brooklyn_bar[0].get_height() + 50, brooklyn_change, fontdict={'fontsize': 13, 'color': 'red'})
plt.text(queens_bar[0].get_x() + 0.1, queens_bar[0].get_height(), queens_change, fontdict={'fontsize': 13, 'color': 'red'})

plt.xticks([1, 2, 4, 5, 7, 8, 10, 11], labels=[
    pandemic_df.index[0], pandemic_df.index[-1],
    pandemic_df.index[0], pandemic_df.index[-1],
    pandemic_df.index[0], pandemic_df.index[-1],
    pandemic_df.index[0], pandemic_df.index[-1]
], rotation=45)

plt.yticks(list(range(0, 4501, 1000)))

plt.xlabel("Dates", fontdict={'fontsize': 13})
plt.ylabel("Median Asking Rental Price (USD)", fontdict={'fontsize': 13})
plt.title("NYC Rental Prices Pre-Pandemic High vs. Sept. 2021", fontdict={'fontsize': 18})

plt.legend()

plt.tight_layout()
plt.show()
```

## What do rental apartment prices look like from the pre-pandemic high to the pandemic low?<a id='second_link'></a>
- Code below, then visualization presented
- *This analysis was conducted in Oct. 2021.*

```python
pandemic_df = df.loc['2019-01-01':'2021-09-01']

# "Fake" Data to set x points
nyc_date_x = [1,2]
manhattan_date_x = [4,5]
brooklyn_date_x = [7,8]
queens_date_x = [10,11]

nyc_prices_y = [pandemic_df['NYC'].max(),pandemic_df['NYC'].min()]
manhattan_prices_y = [pandemic_df['Manhattan'].max(),pandemic_df['Manhattan'].min()]
brooklyn_prices_y = [pandemic_df['Brooklyn'].max(),pandemic_df['Brooklyn'].min()]
queens_prices_y = [pandemic_df['Queens'].max(),pandemic_df['Queens'].min()]

plt.figure(figsize=(8,5))

nyc_bar = plt.bar(nyc_date_x, nyc_prices_y, label='NYC',color ='green',width = 0.4)
manhattan_bar = plt.bar(manhattan_date_x, manhattan_prices_y, label='Manhattan',color ='#0000FF',width = 0.4)
brooklyn_bar = plt.bar(brooklyn_date_x, brooklyn_prices_y, label='Brooklyn',color ='#3333FF',width = 0.4)
queens_bar = plt.bar(queens_date_x, queens_prices_y, label='Queens',color ='#6666FF',width = 0.4)

plt.plot(nyc_date_x,nyc_prices_y,'red',marker='o')
plt.plot(manhattan_date_x,manhattan_prices_y,'red',marker='o')
plt.plot(brooklyn_date_x,brooklyn_prices_y,'red',marker='o')
plt.plot(queens_date_x,queens_prices_y,'red',marker='o')

nyc_change = '- ' + str(round((pandemic_df['NYC'].min() - pandemic_df['NYC'].max()) / pandemic_df['NYC'].max() * -100,1)) + '%'
manhattan_change = '- ' + str(round((pandemic_df['Manhattan'].min() - pandemic_df['Manhattan'].max()) / pandemic_df['Manhattan'].max() * -100,1)) + '%'
brooklyn_change = '- ' + str(round((pandemic_df['Brooklyn'].min() - pandemic_df['Brooklyn'].max()) / pandemic_df['Brooklyn'].max() * -100,1)) + '%'
queens_change = '- ' + str(round((pandemic_df['Queens'].min() - pandemic_df['Queens'].max()) / pandemic_df['Queens'].max() * -100,1)) + '%'

plt.text(nyc_bar[0].get_x() + .6,nyc_bar[0].get_height(),nyc_change,fontdict = {'fontsize' : 13,'color':  'red'})
plt.text(manhattan_bar[0].get_x() + .6,manhattan_bar[0].get_height(),manhattan_change,fontdict = {'fontsize' : 13,'color':  'red'})
plt.text(brooklyn_bar[0].get_x() + .6,brooklyn_bar[0].get_height() + 50,brooklyn_change,fontdict = {'fontsize' : 13,'color':  'red'})
plt.text(queens_bar[0].get_x() + .6,queens_bar[0].get_height(),queens_change,fontdict = {'fontsize' : 13,'color':  'red'})

plt.xticks([1,2,4,5,7,8,10,11],
           labels=[pandemic_df['NYC'].idxmax(),pandemic_df['NYC'].idxmin(),
              pandemic_df['Manhattan'].idxmax(),pandemic_df['Manhattan'].idxmin(),
              pandemic_df['Brooklyn'].idxmax(),pandemic_df['Brooklyn'].idxmin(),
              pandemic_df['Queens'].idxmax(),pandemic_df['Queens'].idxmin()],
          rotation=320)

plt.yticks(list(range(0,4500,500)))
    
plt.xlabel("Dates",fontdict = {'fontsize' : 13})
plt.ylabel("Median Asking Rental Price (USD)",fontdict = {'fontsize' : 13})
plt.title("NYC Rental Prices Pre-Pandemic High vs. Pandemic Low",fontdict = {'fontsize' : 18},y=1.05)

plt.legend()

plt.show()
```

## What are currently the most expensive neighborhoods for a studio apartment in Manhattan and Brooklyn?<a id='third_link'></a>
- Code below, then visualization presented a few cells down
- *This analysis was conducted in Oct. 2021.*

```python
pd.set_option('display.max_rows',10)
pd.set_option('display.max_columns',25)

range1 = [i for i in range(0,3)]
range2 = [i for i in range(123,144)]
usecols = range1 + range2

df_studio = pd.read_csv('medianAskingRent_Studio.csv',usecols=usecols,index_col='areaName')
df_studio.loc[df_studio['areaType'] == 'neighborhood']
df_studio['2020 Avg.'] = df_studio.loc[:,'2020-01':'2020-12'].mean(axis=1)
df_studio['2021 Avg.'] = df_studio.loc[:,'2021-01':'2021-09'].mean(axis=1)
df_studio = df_studio[['Borough','areaType','2020 Avg.','2021 Avg.']]
df_studio = df_studio[df_studio['areaType'] == 'neighborhood']
df_studio.head()
```

```python
studio_bk_21 = df_studio[df_studio['Borough'] == 'Brooklyn']['2021 Avg.'].sort_values(ascending=False).iloc[:5]
studio_bk_21
```

```python
studio_ma_21 = df_studio[df_studio['Borough'] == 'Manhattan']['2021 Avg.'].sort_values(ascending=False).iloc[:5]
studio_ma_21
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1,2,1)
plt.bar(studio_bk_21.index,studio_bk_21)
plt.xticks(rotation=320)
plt.yticks(list(range(0,4500,500)))
plt.xlabel("Neighborhood",fontdict = {'fontsize' : 13})
plt.ylabel("Median Asking Rental Price (USD)",fontdict = {'fontsize' : 13})
plt.title("Most Expensive Brooklyn Neighborhoods for Studio Apartments")

plt.subplot(1,2,2)
plt.bar(studio_ma_21.index,studio_ma_21)
plt.xticks(rotation=320)
plt.yticks(list(range(0,4500,500)))
plt.xlabel("Neighborhood",fontdict = {'fontsize' : 13})
plt.ylabel("Median Asking Rental Price (USD)",fontdict = {'fontsize' : 13})
plt.title("Most Expensive Manhattan Neighborhoods Studio Apartments")

plt.show()
```

## What have apartment prices looked like from Jan. 2012 to Sept. 2021?<a id='fourth_link'></a>
- Code below, then visualization presented
- *This analysis was conducted in Oct. 2021.*

```python
df.index = pd.to_datetime(df.index)

dates_x = df.index
bk_prices_y = df['Brooklyn']
ma_prices_y = df['Manhattan']
qu_prices_y = df['Queens']
nyc_prices_y = df['NYC']

plt.figure(figsize=(12, 6))

plt.plot(dates_x, bk_prices_y, label='Brooklyn', color='#3333FF')
plt.plot(dates_x, ma_prices_y, label='Manhattan', color='#0000FF')
plt.plot(dates_x, qu_prices_y, label='Queens', color='#6666FF')
plt.plot(dates_x, nyc_prices_y, label='NYC', color='green', linewidth=2.5)

plt.xticks(df.index[df.index.month == 1], df.index[df.index.month == 1].strftime('%Y-%m'), rotation=45)

plt.xlabel('Time (Months)', fontdict={'fontsize': 13})
plt.ylabel('Median Asking Rent (USD)', fontdict={'fontsize': 13})
plt.title('NYC Rental Prices (2012 - 2021)', fontdict={'fontsize': 15})

plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```

## What have apartment prices looked like from Jan. 2020 to Sept. 2021?<a id='fifth_link'></a>

- *This analysis was conducted in Oct. 2021.*

```python
dip_df = df.loc['2020-01-01':'2021-09-01']

dates_x = dip_df.index
bk_prices_y = dip_df['Brooklyn']
ma_prices_y = dip_df['Manhattan']
qu_prices_y = dip_df['Queens']
nyc_prices_y = dip_df['NYC']

plt.figure(figsize=(12, 6))

plt.plot(dates_x, bk_prices_y, label='Brooklyn', color='#3333FF')
plt.plot(dates_x, ma_prices_y, label='Manhattan', color='#0000FF')
plt.plot(dates_x, qu_prices_y, label='Queens', color='#6666FF')
plt.plot(dates_x, nyc_prices_y, label='NYC', color='green', linewidth=2.5)

plt.xticks(rotation=45)
plt.yticks(list(range(2500, 4001, 500)))
plt.xlabel("Dates", fontdict={'fontsize': 13})
plt.ylabel("Median Asking Rental Price (USD)", fontdict={'fontsize': 13})
plt.title("NYC Rental Prices (2020 - 2021)", fontdict={'fontsize': 18})

plt.legend(loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```
